```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
```

```python
# Задание 1. Пример построения графика

# Устанавливаем размер и разрешение графика
plt.figure(figsize=(8, 5), dpi=80)
plt.subplot(111)

# Создаем данные для графиков
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

# Строим графики
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

# Настраиваем оси
ax = plt.gca() # Получаем текущие оси
ax.spines['right'].set_color('none')  # Убираем правую грань
ax.spines['top'].set_color('none')  # Убираем верхнюю грань
ax.xaxis.set_ticks_position('bottom')  # Перемещаем отметки оси X внизу
ax.spines['bottom'].set_position(('data', 0))  # Перемещаем ось X на 0
ax.yaxis.set_ticks_position('left')  # Перемещаем отметки оси Y слева
ax.spines['left'].set_position(('data', 0))  # Перемещаем ось Y на 0

# Настраиваем пределы и отметки оси X
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Настраиваем пределы и отметки оси Y
plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.yticks([-1, 1], [r'$-1$', r'$+1$'])

plt.legend(loc='upper left')

# Определяем точку для аннотации
t = 2*np.pi/3
# Линия и точка для косинуса
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=1.5, linestyle= "--")
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')

# Аннотация для синуса
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
    xy=(t, np.sin(t)), xycoords='data',
    xytext=(10, 30), textcoords='offset points', fontsize=16,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad =.2"))

# Линия и точка для синуса
plt.plot([t, t], [0, np.sin(t)], color='red', linewidth=1.5, linestyle= "--")
plt.scatter([t, ], [np.sin(t), ], 50, color ='red')
# Аннотация для косинуса
plt.annotate(r'$cos(\frac{2\pi}{3}) =- \frac{1}{2}$', xy=(t, np.cos(t)),
    xycoords='data', xytext=(-90, -50),
    textcoords='offset points', fontsize=16,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad =.2"))
```

```python
# Задание 2. Рисование нескольких графиков

# Создаем фигуру (окно) с заданными размерами
plt.figure(figsize=(6, 4))
# Определяем сетку 3x3 для размещения подграфиков
G = gridspec.GridSpec(3, 3)

# Создаем первый подграфик, занимающий всю первую строку
axes_1 = plt.subplot(G[0, :])
plt.xticks(()) # Убираем отметки на оси X
plt.yticks(()) # Убираем отметки на оси Y
# Добавляем текст в центр
plt.text(0.5, 0.5, 'Axes 1', ha='center', va='center', size=24, alpha =.5) 

# Создаем второй подграфик, занимающий первую и вторую ячейки второй строки
axes_2 = plt.subplot(G[1, : -1])
plt.xticks(())
plt.yticks(())
plt.text(0.5, 0.5, 'Axes 2', ha='center', va='center', size=24, alpha =.5)

# Создаем третий подграфик, занимающий последние две строки в третьем столбце
axes_3 = plt.subplot(G[1:, -1])
plt.xticks(())
plt.yticks(())
plt.text(0.5, 0.5, 'Axes 3', ha='center', va='center', size=24, alpha =.5)

# Создаем четвертый подграфик, занимающий нижнюю левую ячейку
axes_4 = plt.subplot(G[-1, 0])
plt.xticks(())
plt.yticks(())
plt.text(0.5, 0.5, 'Axes 4', ha='center', va='center', size=24, alpha =.5)

# Создаем пятый подграфик, занимающий нижнюю центральную ячейку
axes_5 = plt.subplot(G[-1, -2])
plt.xticks(())
plt.yticks(())
plt.text(0.5, 0.5, 'Axes 5', ha='center', va='center', size=24, alpha =.5)

plt.tight_layout ()
plt.show()
```

```python
# Задание 4.

n = 256 # Устанавливает количество точек для построения графика.
X = np.linspace(-np.pi, np.pi, n, endpoint = True) 
Y = np.sin(2 * X)
plt.plot(X, Y+1, color='blue', alpha=1.00)
plt.plot(X, Y-1, color='blue', alpha=1.00)

# Заполняет область между линией y=1 и кривой Y+1 там, где Y+1 выше 1, используя синий цвет.
plt.fill_between(X, 1, Y + 1, where=(Y + 1 > 1), color='blue', alpha=0.3)
plt.fill_between(X, -1, Y - 1, where=(Y - 1 < -1), color='red', alpha=0.3)
plt.fill_between(X, Y + 1, 1, where=(Y + 1 < 1), color='blue', alpha=0.3)
plt.fill_between(X, Y - 1, -1, where=(Y - 1 > -1), color='blue', alpha=0.3)
```

```python
# Задание 8. Цветовая карта

def f(x, y):
    return (1 - x/2 + x ** 5 + y ** 3) * np.exp(-x ** 2-y ** 2)
n = 10
x = np.linspace(-3, 3, 4*n)
y = np.linspace(-3, 3, 4*n)
X, Y = np.meshgrid(x, y) # создает двумерные массивы X и Y, представляющие координаты точек в плоскости xy. Каждый элемент массива X содержит значение из массива x, а каждый элемент массива Y содержит значение из массива y.
plt.imshow(f(X, Y), cmap='hot', interpolation='nearest') #  вычисляет значения функции f для каждой точки в массивах X и Y и отображает их в виде изображения с использованием цветовой карты gray и ближайшего соседства для интерполяции пикселов.
plt.colorbar()
```

```python
# Задание 9. Круговые диаграммы

Z = np.random.uniform(0, 1, 20)
Z.sort()

Z[-1] = 0

expl = np.zeros(20)
expl[-2] = 0.1

colors = plt.cm.gray(np.linspace(0, 1, len(Z)))
plt.pie(Z, colors=colors, explode=expl)
```
