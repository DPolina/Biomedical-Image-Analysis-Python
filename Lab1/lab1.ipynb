```python
# Создание массивов

# array(obj, dtype = None, copy = True, order = 'K', subok = False, ndmin = 0)
# Создает массив из любого объекта, который может быть преобразован в массив.
# Например, из списка, кортежа или другого массива.
# dtype: тип элементов массива
# copy: копировать ли данные или использовать ссылку
# order: порядок элементов в памяти
# subok: создать подкласс ndarray, если obj является подклассом ndarray
# ndmin: минимальная размерность массива
arr1 = np.array([1, 2, 3])  # Одномерный массив
arr2 = np.array([[1, 2], [3, 4]])  # Двумерный массив

# zeros(shape, dtype = float, order = 'C')
# Создает массив, заполненный нулями.
# shape: форма массива (кортеж из размеров)
# dtype: тип элементов
# order: порядок элементов в памяти
zeros_arr = np.zeros((3, 3))  # Массив 3x3, заполненный нулями

# zeros_like(a, dtype = None, order = 'K', subok = True, shape = None)
# Создает массив нулей такой же формы и типа, как массив a.
zeros_like_arr = np.zeros_like(arr1)

# ones(shape, dtype = float, order = 'C')
# Создает массив, заполненный единицами.
# Аргументы аналогичны zeros()
ones_arr = np.ones((2, 2))

# ones_like(a, dtype = None, order = 'K', subok = True, shape = None)
# Создает массив единиц такой же формы и типа, как массив a.
ones_like_arr = np.ones_like(arr2)

# empty(shape, dtype = float, order = 'C')
# Создает массив, не инициализируя его значения.
# Содержимое будет произвольным, так как память не очищается.
# Используйте с осторожностью!
empty_arr = np.empty((2, 2))

# empty_like(a, dtype = None, order = 'K', subok = True, shape=None)
# Создает пустой массив такой же формы и типа, как массив a.
empty_like_arr = np.empty_like(arr1)

# Создание массивов с заданным диапазоном значений

# arange([start, ]stop, [step, ]dtype=None)
# Создает массив с равномерно распределенными значениями в заданном диапазоне.
# start: начальное значение (включительно, по умолчанию 0)
# stop: конечное значение (исключительно)
# step: шаг (по умолчанию 1)
arange_arr = np.arange(10)  # Массив от 0 до 9
arange_step_arr = np.arange(0, 1, 0.1)  # Массив от 0 до 1 с шагом 0.1

# linspace(start, stop, num = 50, endpoint = True, retstep = False, dtype = None)
# Создает массив с равномерно распределенными значениями в заданном диапазоне.
# start: начальное значение
# stop: конечное значение
# num: количество элементов в массиве
# endpoint: включить ли конечное значение
linspace_arr = np.linspace(0, 1, 10)  # 10 точек от 0 до 1

# Случайные числа

# random.rand(d0, d1, ..., dn)
# Создает массив случайных чисел в диапазоне [0, 1).
rand_arr = np.random.rand(3, 3)

# random.randn(d0, d1, ..., dn)
# Создает массив случайных чисел из стандартного нормального распределения.
randn_arr = np.random.randn(2, 2)

# Создание массивов из функций

# fromfunction(function, shape, **kwargs)
# Создает массив, вычисляя значения элементов по заданной функции.
def my_function(i, j):
    return i + j
fromfunction_arr = np.fromfunction(my_function, (3, 3))

# fromfile(file, dtype = float, count =- 1, sep = '', offset = 0)
# Создает массив из данных, прочитанных из файла.
# file: имя файла или открытый файл
# dtype: тип элементов
# count: количество элементов для чтения
# sep: разделитель между элементами (если файл содержит текстовые данные)
fromfile_arr = np.fromfile('data.txt', sep = ',')
```

```python
# Task 3. Основные операции

a = np.array([20, 30, 40, 50])
b = np.arange(4)
c = a - b
d = b ** 2
f = 10*np.sin(a)
k = a < 35

print(f"a = {a}")
print(f"b = {b}")
print(f"c = a - b = {c}")
print(f"d = b ** 2 = {d}")
print(f"f = {f}")
print(f"k = {k}")
```

```python
# Task 4. Индексы

# Создаем массив из кубов чисел от 0 до 9
a = np.arange(10) ** 3
print(f"{a}\n")
# Доступ к элементу с индексом 2 (третий элемент)
print(f"{a[2]}")
# Срез массива: элементы с индексами от 2 до 4 (не включая 5)
print(f"{a[2:5]}")
# Изменение элементов с индексами от 0 до 5 (не включая 6) с шагом 2 на -1000
a[:6:2] = -1000
print(f"{a}")
# Разворот массива
print(f"{a[::- 1]}\n")
# Итерация по элементам массива и вычисление кубического корня
for i in a:
    print(i ** (1/3))
```

```python
# Пример индексации многомерного массива

# Создаем массив из 20 элементов с плавающей точкой (от 0 до 19)
# и преобразуем его в матрицу 5х4
b = np.arange(20, dtype = float).reshape(5, 4)

# Выводим элемент на пересечении 3-й строки и 4-го столбца (индексация начинается с 0)
print(b[2, 3])

# Выводим весь второй столбец (индекс 1)
# Диапазон по строкам 0:5 означает, что мы берем все строки
print(b[0:5, 1])

# Эквивалентно предыдущей строке: выводим весь второй столбец
# : означает, что мы берем все элементы по указанному измерению
print(b[:, 1])

# Выводим подматрицу, состоящую из второй и третьей строк и всех столбцов
print(b[1:3, :])

# Выводим последнюю строку матрицы. Эквивалентно b[-1, :]
print(b[-1]) # equals b[-1, :]
```

```python
c = np.array([[[ 0, 1,2],[10,12, 13]],[[100, 101, 102],[110, 112,113]]])
print(c)
print(f"{c[1, ... ]}\n") # equals c[1, :, :] or c[1]
print(f"{c[ ... , 2]}\n") # equals c[:, :, 2]

for element in b.flat:
    print(element)
```
